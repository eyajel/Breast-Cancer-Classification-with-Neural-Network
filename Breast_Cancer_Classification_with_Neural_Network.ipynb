### **Breast** **Cancer Classification with Neural Network**

```python
#  Importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
##  Data Collection and Processing

#  loading the data from sklearn

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

print(breast_cancer_dataset)
```

```python
# Loading the data to a dataframe

data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)

# Print the first 5 rows of the DataFrame

data_frame.head()
```

```python
# Adding the 'target' column to the data frame

data_frame['label'] = breast_cancer_dataset.target

data_frame.tail()
```

```python
# Number of columns and rows in the dataset

data_frame.shape
```

```python
# Getting some information about the data

data_frame.info()
```

```python
# Checking for some missing values

data_frame.isnull().sum()
```

```python
# Statistical measures about the data

data_frame.describe()
```

```python
# Checking the distribution of Target Variable

data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
##  Separating the features and target

X = data_frame.drop(columns='label' , axis=1)
Y = data_frame['label']
print(X)
```

```python
print(Y)
```

```python
##  Spliting the data into training data and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standerdize the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)

print(X_train_std)
```

```python
## Building a Neural Network

# Importing TensorFlow and keras

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers of the neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # one input layer (flatten to convert the input into a single dimensional array and the input shape is 30 because we have 30 features so each feature is a neuron)
    keras.layers.Dense(20,activation='relu'),  # one hidden layer (with 20 neurons)
    keras.layers.Dense(2,activation='sigmoid')  # one ouput layer (with 2 neurons because we have 2 classes as an output : malignant and benign)
])
```

```python
# Compiling the neural network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
## Visualizing accuracy and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc='lower right')
plt.show()
```

```python
## Visualizing accuracy and loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'], loc='upper right')
plt.show()
```

```python
## The accuracy of the model on the test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
print(Y_pred.shape)
print(Y_pred[0])
```

```python
# Converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
